```python
import pandas as pd
import numpy as np
```

```python
nat_gas_df=pd.read_csv('Nat_Gas.csv')
print("\n -- data info ")
print(nat_gas_df.head())
print("\n-- data headings")
print(nat_gas_df.shape)
print("\n Statistical description")
print(nat_gas_df.describe())
print("\n -- number of null values --")
print(nat_gas_df.isnull().sum())
```

```python
#now let's conver date and set index 
#Our date column actually looks like 10/31/2024 , 11,30,2024 etc
#this is actually a string object type
#we have to convert it into datetime objectst that pandas understands

nat_gas_df['Dates']=pd.to_datetime(nat_gas_df['Dates'])
#pd.datetime convers the string to datetime object that pandas undersatnds.
#eg : "10/31/20" → 2020-10-31 00:00:00
#Datetime object allows time bsed operations like resampling,interpolation and plotiing on time axis 
nat_gas_df.set_index('Dates',inplace=True)
#by this way date is set as the default index of the table, Now onwards we can consider as time-series index.
```

```python
#to ensure monthly frequency
nat_gas_df=nat_gas_df.asfreq('M')
#this code will ensures the index follows a monthly schedule
#if there is a gap or missing months pandas will insert those months into the index and fill it wth NaN values.
```

```python
print("\n✅ Converted Dates to datetime and set as index")
print(nat_gas_df.info())
```

```python
#plot tim series
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(nat_gas_df.index, nat_gas_df['Prices'], marker='o', linestyle='-')
plt.title("Natural Gas Prices (Monthly End-of-Month Snapshots)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
#Now we can move forward with seasonal decomposition.
from statsmodels.tsa.seasonal import seasonal_decompose
#now we have to log transform the price for stability.
nat_gas_df['log_price']=np.log(nat_gas_df['Prices'])
#we can perform seasonal decomposition(12 = yearly cycle on monthly data)
decomposition=seasonal_decompose(nat_gas_df['log_price'],model='additive',period=12)
#It splits your series into 3 components:
#Trend ,seasonal, Residual
#Adding components to dataframe

nat_gas_df['Trend']=decomposition.trend
nat_gas_df['Seasonal']=decomposition.seasonal
nat_gas_df['resid']=decomposition.resid

```

```python
#Lets plot decomposition
fig=decomposition.plot()
fig.set_size_inches(12,8)
fig.suptitle("Seasonal Decomposition of Natural Gas Prices (log scale)",fontsize=14)
plt.show()
```

```python
#now we can extrapolate to the next 12 months , that is next year.
from pandas.tseries.offsets import DateOffset

# Future dates: next 12 months after last known point
future_dates = [nat_gas_df.index[-1] + DateOffset(months=x) for x in range(1, 13)]
future_df = pd.DataFrame(index=future_dates, columns=nat_gas_df.columns)

# Combine historical + future placeholder
full_df = pd.concat([nat_gas_df, future_df])

# Seasonal naive forecast:
# Use the average trend from the last year
last_trend = nat_gas_df['Trend'].iloc[-12:].mean()

# Use the last seasonal pattern (12 months)
seasonal_pattern = decomposition.seasonal[-12:].values

# Forecast log prices = last trend + seasonal cycle
future_forecast = last_trend + seasonal_pattern
full_df.loc[future_dates, 'log_price'] = future_forecast

# Convert back from log to price
full_df['Prices_forecast'] = np.exp(full_df['log_price'])

# -----------------------------
# Plot Historical + Forecast
# -----------------------------
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(nat_gas_df.index, nat_gas_df['Prices'], label="Historical", marker="o")
plt.plot(full_df.index, full_df['Prices_forecast'], label="Forecast (Next 12m)", linestyle="--", color="orange")
plt.title("Natural Gas Prices: Historical & 1-Year Extrapolation")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Daily Interpolation Function
def estimate_price(date_input: str) -> float:
    """
    Estimate natural gas price for a given date (string 'YYYY-MM-DD').
    Uses linear interpolation on log-prices between monthly points.
    
    Supported range: Oct 2020 – Sep 2025
    """
    # Convert input to datetime
    date = pd.to_datetime(date_input)
    
    # Check range
    if date < full_df.index.min() or date > full_df.index.max():
        raise ValueError(f"Date out of supported range ({full_df.index.min().date()} – {full_df.index.max().date()}).")
    
    # Interpolation on log scale
    price = np.exp(
        np.interp(
            date.toordinal(),  # numeric representation of date
            full_df.index.to_series().map(lambda x: x.toordinal()),  # x-axis = monthly dates
            full_df['log_price']  # y-axis = log prices
        )
    )
    return round(float(price), 2)


#Example usage

print("Estimated price on 2023-07-15:", estimate_price("2023-07-15"))
print("Estimated price on 2025-05-15:", estimate_price("2025-05-15"))
```
